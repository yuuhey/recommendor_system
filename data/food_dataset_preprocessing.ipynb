```python
def del_duplicates_menu(new_menu,name):
    
    global new_df
    
    a = new_menu[new_menu['메뉴카테고리소분류명']==name].메뉴가격.mean()
    b = new_menu[new_menu['메뉴카테고리소분류명']==name].메뉴가격.median()
    c = new_menu[new_menu['메뉴카테고리소분류명']==name].메뉴가격.mode()[0]
    if c==0: # 최빈값이 0이면 평균과 중앙값으로만 계산
        price = round((a+b)/2, -3)
    else:
        price = round((a+b+c)/3, -3)
    
    new_menu['메뉴가격']= new_menu['메뉴가격'].astype('float')
    
    # mask1 : 메뉴가격에 대표가격이 있는 경우
    # mask2 : 메뉴가격에 대표가격은 없지만 최빈값이 있는 경우
    mask1 = (round(new_menu['메뉴가격'],-3) == price)
    mask2 = (round(new_menu['메뉴가격'],-3) == c)
    
    if len(new_menu[(new_menu['메뉴카테고리소분류명']==name) & (mask1)]) != 0:
#         print("mask1", item, price)
        filtered_df = pd.DataFrame(new_menu[(new_menu['메뉴카테고리소분류명']==name) & (mask1)].iloc[0]).transpose()
        new_df = pd.concat([new_df, filtered_df])
#         display(new_df)
#         new_menu = pd.concat([new_menu[(~mask1)],pd.DataFrame(new_menu[mask1].iloc[0]).transpose()], axis = 0,)
    # elif문 : 대표가격이 메뉴가격에 없고 최빈값이 0이 아니면서 메뉴가격에 존재하는 경우
    elif (c!=0) & (len(new_menu[(new_menu['메뉴카테고리소분류명']==name) & (mask2)]) != 0):
#         print("mask2", item, price)
        filtered_df = pd.DataFrame(new_menu[(new_menu['메뉴카테고리소분류명']==name) & (mask2)].iloc[0]).transpose()
        new_df = pd.concat([new_df, filtered_df])
    else:
        print(item, price)
        
    return new_df
```
